### Task 1: Stacked Chart for Composition of Job Satisfaction Across Age Groups


##### 1. Stacked Chart of Median `JobSatPoints_6` and `JobSatPoints_7` for Different Age Groups


Visualize the composition of job satisfaction scores (`JobSatPoints_6` and `JobSatPoints_7`) across various age groups. This will help in understanding the breakdown of satisfaction levels across different demographics.



```python
# Group data by Age and calculate the median for JobSatPoints_6 and JobSatPoints_7
age_groups = df.groupby('Age')[['JobSatPoints_6', 'JobSatPoints_7']].median()

# Plot the stacked bar chart
age_groups.plot(kind='bar', stacked=True, figsize=(10, 6), color=['#FF6347', '#4682B4'])

# Adding title and labels
plt.title('Median Job Satisfaction Points (6 & 7) Across Age Groups')
plt.xlabel('Age Groups')
plt.ylabel('Median Job Satisfaction Score')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Task 2: Stacked Chart for Compensation and Job Satisfaction by Age Group


##### This stacked chart visualizes the composition of compensation (`ConvertedCompYearly`) and job satisfaction (`JobSatPoints_6`) specifically for respondents aged 30-35.


```python
import matplotlib.pyplot as plt

# Step 1: Filter data for respondents aged 30-35
# We need to filter for ages between '30-34 years old' and '35-44 years old'
age_30_35 = df[df['Age'].isin(['30-34 years old', '35-44 years old'])]

# Step 2: Group by 'Age' and get the median for 'ConvertedCompYearly' and 'JobSatPoints_6'
comp_sat_30_35 = age_30_35.groupby('Age')[['ConvertedCompYearly', 'JobSatPoints_6']].median()

# Step 3: Plot the stacked bar chart
comp_sat_30_35.plot(kind='bar', stacked=True, figsize=(10, 6), color=['#FF6347', '#4682B4'])

# Adding title and labels
plt.title('Compensation and Job Satisfaction (Age 30-35)')
plt.xlabel('Age')
plt.ylabel('Median Values')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Task 3: Comparing Data Using Stacked Charts


##### 1. Stacked Chart of Preferred Databases by Age Group




Visualize the top databases that respondents from different age groups wish to learn. Create a stacked chart to show the proportion of each database in each age group.


```python
import pandas as pd
import matplotlib.pyplot as plt

# Ensure the column 'DatabaseAdmired' is of string type (if needed)
df['DatabaseAdmired'] = df['DatabaseAdmired'].astype(str)

# Apply str.split to the 'DatabaseAdmired' column and explode the values by age group
age_grouped_databases = df.groupby('Age')['DatabaseAdmired'].apply(lambda x: x.str.split(';').explode()).reset_index()

# Get the value counts of databases per age group
database_counts_by_age = age_grouped_databases.groupby(['Age', 'DatabaseAdmired']).size().unstack(fill_value=0)

# Plot the stacked bar chart
database_counts_by_age.plot(kind='bar', stacked=True, figsize=(10, 6), colormap='tab20')

plt.legend(title='Database Admired', bbox_to_anchor=(1.05, 1), loc='upper left')
# Adding title and labels
plt.title('Preferred Databases by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Number of Respondents')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

##### 2. Stacked Chart of Employment Type by Job Satisfaction


Analyze the distribution of employment types within each job satisfaction level using a stacked chart. This will provide insights into how employment types are distributed across various satisfaction ratings.


```python
# Ensure the 'Employment' column is clean (strip spaces, convert to string if necessary)
df['Employment'] = df['Employment'].str.strip().astype(str)

# Check for any missing values or irregularities in 'Employment' and 'JobSatPoints_6'
# print(df[['Employment', 'JobSatPoints_6']].isnull().sum())

# Split 'Employment' data by ';' and explode the list for individual rows
employment_split = df.groupby('JobSatPoints_6')['Employment'].apply(lambda x: x.str.split(';').explode()).reset_index()

# Check the result after exploding to ensure correctness
# print("Exploded Employment data:\n", employment_split.head())

# Get the value counts for employment types by job satisfaction level
employment_counts_by_sat = employment_split.groupby(['JobSatPoints_6', 'Employment']).size().unstack(fill_value=0)

# Check the resulting DataFrame to confirm
# print("Employment counts by job satisfaction:\n", employment_counts_by_sat)

# Plot the stacked bar chart
ax = employment_counts_by_sat.plot(kind='bar', stacked=True, figsize=(30, 6), colormap='tab20')

# Adjust the legend position
plt.legend(title='Employment Type', bbox_to_anchor=(1.05, 1), loc='upper left')

# Adding title and labels
plt.title('Employment Type Distribution by Job Satisfaction Level')
plt.xlabel('Job Satisfaction Level')
plt.ylabel('Number of Respondents')
plt.xticks(rotation=45)

# Ensure the layout fits well
plt.tight_layout()

# Show the plot
plt.show()
```

### Task 4: Exploring Technology Preferences Using Stacked Charts


##### 1. Stacked Chart for Preferred Programming Languages by Age Group


Analyze how programming language preferences (`LanguageAdmired`) vary across age groups.


```python
##Write your code here
# Ensure the 'LanguageAdmired' column is of string type (if needed)
df['LanguageAdmired'] = df['LanguageAdmired'].astype(str)

# Split the programming languages respondents admire by age group
age_grouped_languages = df.groupby('Age')['LanguageAdmired'].apply(lambda x: x.str.split(';').explode()).reset_index()

# Get value counts of programming languages by age group
language_counts_by_age = age_grouped_languages.groupby(['Age', 'LanguageAdmired']).size().unstack(fill_value=0)

# Plot the stacked bar chart
ax = language_counts_by_age.plot(kind='bar', stacked=True, figsize=(10, 6), colormap='tab20')

# Adjust legend position
plt.legend(title='Programming Languages', bbox_to_anchor=(1.05, 1), loc='upper left')

# Adding title and labels
plt.title('Preferred Programming Languages by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Number of Respondents')
plt.xticks(rotation=45)

# Ensure the layout fits well
plt.tight_layout()

# Show the plot
plt.show()
```

##### 2. Stacked Chart for Technology Adoption by Employment Type


Explore how admired platforms (`PlatformAdmired`) differ across employment types (e.g., full-time, freelance)


```python
import pandas as pd
import matplotlib.pyplot as plt

# Ensure the 'PlatformAdmired' column is of string type (if needed)
df['PlatformAdmired'] = df['PlatformAdmired'].astype(str)

# Step 1: Split and explode both 'Employment' and 'PlatformAdmired' columns
df_exploded = df.copy()

# Split and explode 'Employment' column
df_exploded['Employment'] = df_exploded['Employment'].str.split(';')
df_exploded = df_exploded.explode('Employment')

# Split and explode 'PlatformAdmired' column
df_exploded['PlatformAdmired'] = df_exploded['PlatformAdmired'].str.split(';')
df_exploded = df_exploded.explode('PlatformAdmired')

# Step 2: Get value counts of admired platforms by the exploded employment type
platform_counts_by_employment = df_exploded.groupby(['Employment', 'PlatformAdmired']).size().unstack(fill_value=0)

# Step 3: Plot the stacked bar chart
ax = platform_counts_by_employment.plot(kind='bar', stacked=True, figsize=(10, 6), colormap='tab20')

# Adjust legend position
plt.legend(title='Admired Platforms', bbox_to_anchor=(1.05, 1), loc='upper left')

# Adding title and labels
plt.title('Technology Adoption by Employment Type')
plt.xlabel('Employment Type')
plt.ylabel('Number of Respondents')
plt.xticks(rotation=45, ha = "right")

# Ensure the layout fits well
plt.tight_layout()

# Show the plot
plt.show()
```
